# **Q1. What is Two-tailed and One-tailed t-Test?**
🔹 One-Tailed t-Test

A one-tailed t-test is used when the alternative hypothesis specifies the direction of the effect (either greater than or less than).

📌 Types:

Right-tailed test → checks if value is greater

H₀: μ ≤ 19

H₁: μ > 19

Left-tailed test → checks if value is smaller

H₀: μ ≥ 19

H₁: μ < 19

🔹 Use Case:

When we are only interested in one direction of change.

🔹 Two-Tailed t-Test

A two-tailed test is used when we want to check whether the value is different (either greater or smaller).

📌 Hypothesis:

H₀: μ = 19

H₁: μ ≠ 19

🔹 Characteristics:

Rejection region is divided into two parts (both ends)

More strict than one-tailed test

🔹 Key Difference:
Feature	One-Tailed	Two-Tailed
Direction	One side	Both sides
Hypothesis	μ > or μ <	μ ≠
Critical Region	One side	Two sides
Usage	Specific direction	Any difference

# **Q2. Explain Type I and Type II Errors with Example**

In hypothesis testing, errors can occur while making decisions.

🔹 Type I Error (α Error)

Rejecting H₀ when it is actually true

📌 Meaning:

False alarm

📌 Example:

H₀: Student average age = 19

We reject H₀, but actually the average is 19

👉 So we made a wrong rejection

🔹 Type II Error (β Error)

Accepting H₀ when it is actually false

📌 Meaning:

Missed detection

📌 Example:

H₀: Average age = 19

Actual average ≠ 19

But we accept H₀

👉 So we failed to detect difference

🔹 Summary Table:
Error Type	Meaning	Symbol
Type I	Reject true H₀	α
Type II	Accept false H₀	β

# **Q4. Explain input and output characteristics of stats.ttest_rel()**

stats.ttest_rel() is used to perform a paired t-test.

🔹 Purpose:

It compares two related samples to check if their means are significantly different.

🔹 Input Parameters:

📌 Inputs:

a → First set of observations

b → Second set of observations

🔹 Conditions:

Both samples must be:

Same size

Related (paired data)

Example:

Before and after marks

Pre-test and post-test results

📌 Outputs:

t-statistic

Measures difference between paired samples

p-value

Helps decide whether difference is significant

🔹 Interpretation:

If p-value ≤ α → Reject H₀

If p-value > α → Accept H₀

```python
from scipy import stats

before = [10, 12, 14, 16, 18]
after  = [12, 14, 15, 18, 20]

t_stat, p_value = stats.ttest_rel(before, after)

print(t_stat, p_value)
```

# **Q5. Function in stats module for identifying power of statistical test**

In scipy.stats, there is no direct built-in function for power calculation like t-test functions.

Instead, power is calculated using:

👉 statsmodels library

```python
from statsmodels.stats.power import TTestPower
```

```python
from statsmodels.stats.power import TTestPower

analysis = TTestPower()

power = analysis.power(effect_size=0.5, nobs=10, alpha=0.05)

print("Power:", power)
```

🔹 Output Range:

Power value ranges between 0 and 1

📌 Meaning:

0 → No ability to detect effect

1 → Perfect detection

Significance of Output:
Power Value	Meaning
< 0.5	Poor test
~0.8	Good (standard)
> 0.9	Very strong test
        
🔹 Importance:

Helps in selecting sample size

Improves test reliability

Reduces Type II error
