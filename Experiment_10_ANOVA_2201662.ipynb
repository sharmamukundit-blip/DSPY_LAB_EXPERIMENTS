# **Q1. Explain input and output characteristics for f_oneway function from stats library**

f_oneway() function (from scipy.stats) is used to perform one-way ANOVA (Analysis of Variance).

🔹 Purpose:

It tests whether the means of two or more independent groups are significantly different.

🔹 Syntax:

```python
from scipy.stats import f_oneway

f_stat, p_value = f_oneway(group1, group2, group3, ...)
```

🔹 Input Characteristics:

Multiple Groups

Accepts 2 or more groups (arrays/lists)

Independent Samples

Each group should be independent of others

Numeric Data

Data should be continuous (e.g., marks, height)

Same Type of Measurement

All groups must measure same variable

# **Q2. Explain results of sm.stats.anova_lm function**

anova_lm() (from statsmodels) provides a detailed ANOVA table.

🔹 Purpose:

Used to analyze variance in regression models and compare multiple groups.

🔹 Example:

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# Create a sample DataFrame 'df' for demonstration
data = {
    'marks': [85, 90, 78, 92, 88, 75, 80, 95, 82, 87, 79, 91, 84, 86, 93],
    'group': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C']
}
df = pd.DataFrame(data)

model = ols('marks ~ group', data=df).fit()
result = sm.stats.anova_lm(model)

print(result)
```

🔹 Output Table Explanation:
Column	Meaning
sum_sq	Sum of squares (variation)
df	Degrees of freedom
F	F-statistic
PR(>F)	p-value
🔹 Explanation:

sum_sq → variation due to factor

df → number of independent values

F value → ratio of variances

PR(>F) → probability value

🔹 Interpretation:

If PR(>F) ≤ 0.05 → significant difference exists

If PR(>F) > 0.05 → no significant difference

4. Reducing Error

Avoids multiple testing errors

5. Business Decision Making

Helps in choosing best strategy among alternatives

🔹 Applications:

Marketing → compare campaign performance

# **Q4. When to use One-Way ANOVA? (with real-world case)**
🔹 When to Use:

When comparing means of 3 or more groups

When there is only one independent variable (factor)

🔹 Example (Real-world):
🎓 Case: Student Performance

Compare marks of students from:

IT department

CS department

Mechanical department

👉 Factor: Department
👉 Output: Marks

🔹 Hypothesis:

H₀: All group means are equal

H₁: At least one mean is different

🔹 Python Example:

```python
from scipy.stats import f_oneway

it = [70, 75, 80]
cs = [60, 65, 70]
mech = [55, 60, 58]

f_stat, p_value = f_oneway(it, cs, mech)

print(p_value)
```

🔹 Conclusion:

If p ≤ 0.05 → At least one group is different

# **Q5. When to use Two-Way ANOVA? (with real-world case)**
🔹 When to Use:

When there are two independent variables (factors)

To study:

Individual effect of each factor

Interaction effect between factors

🔹 Example (Real-world):
🏫 Case: Student Performance

Factors:

Teaching Method (Online / Offline)

Gender (Male / Female)

Output:

Marks

🔹 What it analyzes:

Effect of teaching method

Effect of gender

Combined effect (interaction)

🔹 Hypothesis:

H₀₁: No effect of factor 1

H₀₂: No effect of factor 2

H₀₃: No interaction effect

🔹 Python Example:

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# Create a sample DataFrame 'df' for two-way ANOVA demonstration
# It needs 'marks', 'method', and 'gender' columns
data = {
    'marks': [85, 90, 78, 92, 88, 75, 80, 95, 82, 87, 79, 91, 84, 86, 93,
              70, 75, 80, 65, 82, 68, 72, 77, 85, 79, 73, 76, 81, 89, 83],
    'method': ['Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online',
               'Offline', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline',
               'Online', 'Online', 'Online', 'Online', 'Online', 'Offline', 'Offline', 'Offline', 'Offline', 'Offline'],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

model = ols('marks ~ method + gender + method:gender', data=df).fit()
anova_table = sm.stats.anova_lm(model)

print(anova_table)
```

🔹 Interpretation:

Check p-values for:

method

gender

interaction

If p ≤ 0.05 → factor has significant effect
