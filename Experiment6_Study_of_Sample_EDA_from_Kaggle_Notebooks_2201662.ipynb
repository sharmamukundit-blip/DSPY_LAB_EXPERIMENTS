##**2] In IRIS data set which non graphical EDA is performed ?**
##Non-Graphical EDA Performed on Iris Dataset

For the Iris Dataset, non-graphical exploratory data analysis involves analyzing the dataset using statistical and tabular methods without using any visual plots.

### **Techniques Performed**

1. **Summary Statistics**
   Calculation of mean, median, minimum, maximum, standard deviation, and variance for all numerical features (sepal length, sepal width, petal length, petal width) to understand central tendency and dispersion.

2. **Class Distribution Analysis**
   Counting the number of instances in each class (Setosa, Versicolor, Virginica) to check whether the dataset is balanced. The dataset contains 50 samples in each class.

3. **Missing Value Analysis**
   Checking for null or missing values in the dataset. The Iris dataset does not contain any missing values.

4. **Data Type Identification**
   Identifying the type of each feature:

   * Input features: numerical (continuous)
   * Target variable (species): categorical (nominal)

5. **Correlation Analysis**
   Computing the correlation matrix to identify relationships between features. For example, petal length and petal width show strong positive correlation.

6. **Range Calculation**
   Determining the difference between maximum and minimum values for each feature to understand the spread of data.

7. **Quartiles and Interquartile Range (IQR)**
   Calculating Q1, Q2 (median), and Q3, and deriving IQR (Q3 − Q1) to understand data distribution and variability.

8. **Outlier Detection (Numerical Method)**
   Using the IQR method to detect outliers by defining lower and upper bounds.

9. **Covariance Analysis**
   Measuring how two variables change together to understand the direction of relationships.

10. **Unique Value Analysis**
    Identifying unique values in the target column, confirming there are three distinct species.

11. **Feature-wise Aggregation**
    Computing statistics (mean, min, max) for each feature grouped by species to compare class characteristics.

12. **Duplicate Record Check**
    Verifying whether duplicate rows exist in the dataset. No duplicates are present.

13. **Skewness Measurement**
    Checking whether the distribution of numerical features is symmetric or skewed.

14. **Coefficient of Variation**
    Measuring relative variability using the ratio of standard deviation to mean.

15. **Frequency Distribution (Tabular Form)**
    Creating frequency tables to observe how values are distributed across intervals.


### **Conclusion**

Non-graphical EDA on the Iris dataset provides a complete statistical understanding of the data, including distribution, relationships, variability, and data quality, without using visualization techniques.


## **3] In  Air Quality data set which graph is used to perform multivariate analysis?**

---

## **Multivariate Analysis in Air Quality Dataset**

For the Air Quality Dataset, multivariate analysis is used to study the relationship among multiple variables such as pollutant concentrations (PM2.5, PM10, NO₂, CO, SO₂, O₃) and meteorological parameters (temperature, humidity, wind speed).

---

## **Graphs Used for Multivariate Analysis**

### **1) Scatter Plot Matrix (Pair Plot)**

* A collection of scatter plots showing pairwise relationships between multiple variables.
* Each variable is plotted against every other variable.
* The diagonal often shows distribution (histogram or density).
* **Purpose:**

  * Identify correlations between pollutants
  * Detect patterns, clusters, and outliers
* **Example:** Relationship between PM2.5, NO₂, CO, and temperature.

---

### **2) Correlation Heatmap**

* A color-coded matrix representing correlation coefficients between variables.
* Values range from **-1 to +1**:

  * +1 → strong positive correlation
  * -1 → strong negative correlation
  * 0 → no correlation
* **Purpose:**

  * Quickly identify highly correlated pollutants
  * Detect multicollinearity
* **Example:** PM2.5 often shows strong correlation with PM10.

---

### **3) 3D Scatter Plot**

* Represents three variables on X, Y, and Z axes.
* **Purpose:**

  * Visualize interaction among three features simultaneously
* **Example:** PM2.5 vs Temperature vs Humidity

---

### **4) Parallel Coordinates Plot**

* Each variable is represented as a vertical axis.
* Each observation is drawn as a line connecting values across axes.
* **Purpose:**

  * Compare multiple variables for each observation
  * Identify patterns across many dimensions

---

### **5) Bubble Chart**

* Extension of scatter plot:

  * X-axis → one variable
  * Y-axis → second variable
  * Bubble size → third variable
* **Purpose:**

  * Represent three variables in a single plot
* **Example:** PM2.5 vs NO₂ with CO as bubble size

---

### **6) Multivariate Time Series Plot**

* Used when dataset includes time component.
* Multiple variables plotted over time in one graph.
* **Purpose:**

  * Analyze trends and seasonal variations of pollutants

---

### **7) PCA Plot (Principal Component Analysis)**

* Reduces multiple variables into 2 or 3 principal components.
* **Purpose:**

  * Simplify high-dimensional data
  * Identify major contributing factors

---

## **Conclusion**

Multivariate analysis of the Air Quality dataset is commonly performed using Scatter Plot Matrix and Correlation Heatmap as primary tools, along with advanced visualizations like 3D Scatter Plot, Parallel Coordinates Plot, Bubble Chart, Time Series Plot, and PCA Plot. These techniques help in understanding relationships, dependencies, and patterns among multiple environmental variables.
