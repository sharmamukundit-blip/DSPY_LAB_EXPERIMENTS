# **Q1. When do we go for Chi-Square Test?**

Chi-square test is a non-parametric statistical test used to analyze categorical data. It is mainly used when we want to study the relationship between variables or check how observed data differs from expected data.

🔹 When to Use Chi-Square Test:
1. For Categorical Data

When data is in the form of categories (e.g., gender, yes/no, pass/fail)

2. To Compare Observed vs Expected Frequencies

When we want to check whether actual results match expected results

3. When Sample Size is Large

Works best when sample size is sufficiently large

4. When Variables are Independent

Observations should not influence each other

🔹 Types of Chi-Square Tests:

Goodness of Fit Test

Checks if data follows expected distribution

Test of Independence

Checks relationship between two variables

🔹 Example:

Checking if student grades follow expected distribution

Checking if gender and course selection are related

```python
from scipy.stats import chi2_contingency

data = [[30, 20],
        [25, 25]]

chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square:", chi2)
print("p-value:", p)
```

🔹 Interpretation:

If p ≤ 0.05 → Variables are dependent

If p > 0.05 → Variables are independent

# **Q3. Role of Critical Value and p-value in Chi-Square Test**
🔹 Critical Value:

It is a threshold value obtained from chi-square table

Depends on:

Degrees of freedom

Significance level (α)

📌 Role:

If χ² calculated > critical value → Reject H₀

🔹 p-value:

Probability of obtaining result assuming H₀ is true

📌 Role:

If p ≤ α → Reject H₀

If p > α → Accept H₀

🔹 Difference:
Feature	Critical Value	p-value
Source	Table	Computed
Decision	Compare χ²	Compare with α
Use	Traditional	Modern

🔹 Conclusion:

Both methods help in decision-making, but p-value is more commonly used in Python.

# **Q4. Can we use Chi-Square test to show correlation between two features?**
🔹 Answer:

Yes, Chi-square test can be used to check association (relationship) between two categorical variables, but it does not measure strength like correlation coefficient.

🔹 Explanation:

It tells whether variables are:

Independent

Dependent (associated)

🔹 Hypothesis:

H₀: No relationship between variables

H₁: Relationship exists

🔹 Interpretation:

After test:

If p ≤ 0.05
→ Reject H₀
→ Relationship exists

If p > 0.05
→ Accept H₀
→ No relationship

🔹 Important Note:

It shows existence of association, not:

Strength
