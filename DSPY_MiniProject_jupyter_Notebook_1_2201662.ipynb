# **Titanic Dataset** - **Data **selection**,Cleaning and** **Wrangling**

The Titanic dataset contains information about passengers who were aboard the Titanic ship.
Each row in the dataset represents an individual passenger, and the columns describe different attributes such as survival status, passenger class, gender, age, and fare.

This dataset is widely used for data analysis and machine learning tasks as it contains a mix of numerical and categorical variables and also includes missing values, making it suitable for practicing data cleaning and preprocessing techniques.

The dataset consists of multiple rows and columns representing passenger data.

The columns include both numerical and categorical data types:
- Numerical columns: Age, Fare
- Categorical columns: Sex
- Discrete numerical column: Pclass

From the dataset overview, it is observed that some columns contain missing values. In particular, the Age column has missing entries, which need to be handled before performing further analysis.

For the purpose of analysis, only the most relevant columns are selected from the dataset. This helps in reducing complexity and focusing only on meaningful features.

The selected columns are:
- Survived: This is the target variable indicating whether a passenger survived (1) or not (0).
- Pclass: This represents the passenger class (1st, 2nd, 3rd), which can be considered as a proxy for socio-economic status.
- Sex: This indicates the gender of the passenger.
- Age: This represents the age of the passenger.
- Fare: This indicates the ticket price paid by the passenger.

These features are chosen because they are most likely to influence survival and are useful for analysis.

Missing values are an important aspect of data preprocessing. If not handled properly, they can lead to incorrect analysis results.

Upon checking the dataset, it is found that the Age column contains missing values. Since Age is a numerical variable, these missing values need to be handled appropriately.

The missing values in the Age column are handled by replacing them with the mean (average) of the Age column.

This approach is chosen because:
- The mean is a good representation of the central tendency for numerical data.
- It ensures that no data is lost by removing rows.
- It maintains the overall distribution of the data.

Thus, all missing values in the Age column are replaced with the average age of passengers.

Data cleaning is an essential step in preprocessing to improve data quality.

The following cleaning steps are performed:

1. Duplicate Removal:
Duplicate rows are removed from the dataset to ensure that each observation is unique and does not bias the analysis.

2. Categorical Encoding:
The 'Sex' column is a categorical variable containing text values ('male', 'female').
To make it suitable for analysis and machine learning, it is converted into numerical form:
- male → 0
- female → 1

This process is known as encoding and helps in simplifying computations.

**working:**

```python
import pandas as pd
import numpy as np
```

In this notebook, we perform data cleaning and preprocessing on the Titanic dataset.
The goal is to prepare the dataset for further analysis by handling missing values,
removing unnecessary data, and converting categorical variables into numerical format.

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

The dataset contains information about passengers such as survival status, age, gender, class, and fare.
Each row represents an individual passenger.

```python
df.shape
```

The dataset consists of multiple rows and columns representing passenger details.

```python
df.columns
```

```python
df.info()
```

From the dataset information, we observe that:
- The dataset contains both numerical and categorical data.
- The Age column has missing values.
- Other columns are mostly complete.

```python
df.describe()
```

The statistical summary shows the distribution of numerical columns such as Age and Fare,
including mean, minimum, and maximum values.

```python
df = df[['Survived','Pclass','Sex','Age','Fare']]
df.head()
```

Only relevant columns are selected to simplify the analysis:
- Survived (target variable)
- Pclass (passenger class)
- Sex (gender)
- Age (age of passenger)
- Fare (ticket price)

```python
df.isnull().sum()
```

The Age column contains missing values which must be handled before further analysis.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

Missing values in the Age column are replaced with the mean value.
This ensures that no data is lost and maintains the distribution of the dataset.

```python
df.drop_duplicates(inplace=True)
```

Duplicate rows are removed to avoid redundancy and ensure data quality.

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

The categorical variable 'Sex' is converted into numerical format:
- male = 0
- female = 1

This makes the dataset suitable for analysis and machine learning.

```python
df.info()
df.head()
```

```python
df.to_csv("cleaned_Titanic-Dataset.csv", index=False)
```

After performing data cleaning and preprocessing steps, the dataset is now clean and structured.

- All missing values have been handled.
- Duplicate entries have been removed.
- Categorical variables have been converted into numerical format.

The dataset is now ready for exploratory data analysis (EDA).
