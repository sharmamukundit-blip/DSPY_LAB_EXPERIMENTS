# **Titanic Dataset - Exploratory Data Analysis**

Exploratory Data Analysis (EDA) is an important step in any data science or machine learning project. It involves analyzing and summarizing datasets to understand their main characteristics before applying any modeling techniques.

EDA helps in discovering patterns, detecting anomalies, testing assumptions, and checking data quality using statistical summaries and visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_Titanic-Dataset.csv")
```

In this notebook, we perform Exploratory Data Analysis (EDA) on the cleaned Titanic dataset.
The aim is to understand patterns, relationships, and key insights from the data using both graphical and non-graphical methods.

# **GRAPHICAL ANALYSIS 1 (UNIVARIATE)**

```python
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

Type of EDA: Graphical Univariate Analysis

What I am analyzing:
I am analyzing the distribution of passenger ages.

What I observe:
Most passengers fall within the age range of approximately 20 to 40 years.
There are fewer passengers in the extreme age groups such as very young children and elderly individuals.

Conclusion:
The dataset is skewed towards younger passengers, indicating that the majority of travelers were adults.

# **GRAPHICAL ANALYSIS 2 (BIVARIATE)**

```python
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Fare Distribution by Class and Survival")
plt.show()
```

Type of EDA: Graphical Bivariate Analysis

What I am analyzing:
I am analyzing the relationship between gender (Sex) and survival (Survived).

What I observe:
Female passengers have a significantly higher survival rate compared to male passengers.

Conclusion:
Gender played a major role in survival, suggesting that females were given priority during rescue operations.

# **NON-GRAPHICAL ANALYSIS 1 (BIVARIATE)**

```python
df.groupby('Sex')['Survived'].mean()
```

Type of EDA: Non-Graphical Bivariate Analysis

What I am analyzing:
I am analyzing the average survival rate grouped by gender.

What I observe:
The survival rate for females is much higher than that of males.

Conclusion:
This confirms that gender has a strong influence on survival probability.

# **NON-GRAPHICAL ANALYSIS 2 (MULTIVARIATE)**

```python
df.groupby(['Pclass','Sex'])['Survived'].mean()
```

Type of EDA: Non-Graphical Multivariate Analysis

What I am analyzing:
I am analyzing how passenger class (Pclass) and gender (Sex) together affect survival.

What I observe:
Passengers in higher classes (1st class) have higher survival rates.
Additionally, females across all classes tend to survive more than males.

Conclusion:
Survival is influenced by multiple factors, particularly class and gender combined.

# **GRAPHICAL ANALYSIS 3 (MULTIVARIATE BONUS)**

```python
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Fare Distribution by Class and Survival")
plt.show()
```

Type of EDA: Graphical Multivariate Analysis

What I am analyzing:
I am analyzing how fare and passenger class relate to survival.

What I observe:
Passengers who paid higher fares (usually in higher classes) tend to have higher survival rates.

Conclusion:
Economic status (indicated by fare and class) played a significant role in survival.

# **GRAPHICAL ANALYSIS 4 (MULTIVARIATE BONUS)**

```python
sns.histplot(data=df, x='Fare', hue='Pclass', bins=30)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

Type: Graphical Multivariate Analysis

What I am analyzing:
I am analyzing how fare varies across different passenger classes.

What I observe:
Higher class passengers tend to pay higher fares, while lower class passengers pay lower fares.

Conclusion:
Fare is strongly related to passenger class and reflects socio-economic differences.

# **GRAPHICAL ANALYSIS 5 (MULTIVARIATE BONUS)**

```python
sns.stripplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Fare vs Class by Survival")
plt.show()
```

Type: Graphical Multivariate Analysis

What I am analyzing:
I am analyzing individual data points of fare across passenger classes and survival.

What I observe:
Higher fares are mostly associated with first-class passengers, and many of them survived.

Conclusion:
Passengers with higher fares had better chances of survival.

# **conclusion:**

From the exploratory data analysis, it is evident that survival on the Titanic was influenced by multiple factors.

Key findings include:
- Gender had a strong impact, with females having higher survival rates.
- Passenger class also influenced survival, with higher-class passengers more likely to survive.
- Age distribution shows that most passengers were adults.
- Fare indicates socio-economic status, which also affected survival.

Overall, survival was not dependent on a single factor but a combination of demographic and socio-economic variables.
