# **Q1. What is the need for Hypothesis Testing?**

Hypothesis testing is a fundamental statistical tool used to make decisions about a population based on sample data. In practical situations, studying the entire population is often impossible due to time, cost, or feasibility constraints
. Therefore, hypothesis testing provides a scientific framework to draw conclusions using limited data.

🔹 Detailed Need:
1. Decision Making Under Uncertainty

In real-life scenarios, data is often uncertain or variable. Hypothesis testing helps in making reliable decisions despite this uncertainty.

2. Testing Claims and Assumptions

It is used to verify claims such as:

Average age of students is 19

A machine produces standard output

A medicine is effective

3. Scientific Validation

Hypothesis testing provides a structured and mathematical way to validate theories and research findings.

4. Comparison Between Groups

It helps compare:

Two different populations

Before and after results

Control and experimental groups

5. Minimizing Errors

It reduces chances of wrong conclusions by controlling:

Type I Error (false rejection)

Type II Error (false acceptance)

6. Widely Applicable

Used in:

Engineering

Medical research

Business analytics

Social sciences

🔹 Example:

Suppose we assume that the average age of TE IT students is 19. Hypothesis testing helps determine whether this assumption is statistically valid or just a coincidence based on sample data.

# **Q2. Explain steps in Hypothesis Testing with neat diagram.**

Hypothesis testing follows a systematic and logical procedure to test a given assumption.

🔹 Detailed Steps:
Step 1: Formulation of Hypotheses

Null Hypothesis (H₀): Assumes no difference or effect

Alternative Hypothesis (H₁): Assumes there is a difference

Example:
H₀: μ = 19
H₁: μ ≠ 19

Step 2: Selection of Significance Level (α)

It defines the probability of rejecting a true null hypothesis

Common values: 0.05, 0.01

Step 3: Selection of Appropriate Test

Depending on:

Sample size

Known/unknown population variance

Types:

z-test (large sample)

t-test (small sample)

Step 4: Calculation of Test Statistic

A numerical value is calculated using sample data:

z-value or t-value

Step 5: Determination of Critical Region / p-value

p-value indicates probability of observing result under H₀

Compared with α

Step 6: Decision Making

If p-value ≤ α → Reject H₀

If p-value > α → Accept H₀

Step 7: Conclusion

Interpret result in context of problem.

# **Q3. What is Student t-distribution?**

Student’s t-distribution is a probability distribution used in hypothesis testing when the sample size is small and population standard deviation is unknown.

🔹 Detailed Explanation:

It was developed by William Sealy Gosset under the pseudonym “Student”.

When the sample size is small, the normal distribution is not reliable. In such cases, t-distribution is used as it accounts for additional uncertainty.

🔹 Characteristics:

Bell-Shaped and Symmetrical
Similar to normal distribution but more spread out

Heavier Tails
More probability in extreme values

Depends on Degrees of Freedom
df = n - 1

Approaches Normal Distribution
As sample size increases

🔹 Importance:

Used in small sample testing

More accurate when population variance is unknown

Forms the basis of t-tests

🔹 Applications:

One-sample t-test

Independent sample t-test

Paired t-test

# **Q5. Design hypothesis testing for average age = 19**
🔹 Step 1: Problem Statement

Check whether the average age of TE IT students is 19 or not.

🔹 Step 2: Hypotheses

H₀: μ = 19

H₁: μ ≠ 19

🔹 Step 3: Significance Level

α = 0.05

🔹 Step 4: Test Selection

One-sample t-test

Reason:

Small sample

Unknown population standard deviation

Step 5: Sample Data

```python
ages = [18, 19, 20, 19, 18, 21, 19, 20, 18, 19]
```

Step 6: Calculation

```python
import numpy as np

ages = np.array([18,19,20,19,18,21,19,20,18,19])

mean = np.mean(ages)
std = np.std(ages, ddof=1)
n = len(ages)

t_value = (mean - 19) / (std / np.sqrt(n))

print("Mean:", mean)
print("t-value:", t_value)
```

🔹 Step 7: Decision Rule

Reject H₀ if p ≤ 0.05

Accept H₀ if p > 0.05

# **Q6. Check acceptance of H₀ using Python**
🔹 Using SciPy:

```python
from scipy import stats

ages = [18,19,20,19,18,21,19,20,18,19]

t_stat, p_value = stats.ttest_1samp(ages, 19)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

🔹 Decision Making:

```python
alpha = 0.05

if p_value <= alpha:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

Final Interpretation:

If p-value > 0.05
→ No significant difference
→ Average age is 19

If p-value ≤ 0.05
→ Significant difference
→ Average age is not 19
